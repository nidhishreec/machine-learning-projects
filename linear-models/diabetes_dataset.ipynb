IMPORT LIBRARIES

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

CREATE DATAFRAME

```python
df = pd.DataFrame({
    'Age (years)': [25, 30, 35, 40, 45, 50, 55, 60],
    'Glucose Level (mg/dL)': [85, 90, 100, 110, 120, 135, 145, 160],
    'Diabetes Risk (%)': [5, 10, 18, 28, 40, 55, 70, 85]
})

df.head()
```

Define features (X) and target (y)

```python
X = df[['Age (years)','Glucose Level (mg/dL)']]
y = df['Diabetes Risk (%)']           
```

```python
plt.figure(figsize=(4, 4))
plt.plot(X,y)
plt.show()
```

Observation: The graph shows a linear relationship

Split into training and testing sets

```python
Xtrain,Xtest,Ytrain,Ytest = train_test_split(X, y, train_size=0.8, random_state=42)
```

Create and train Linear Regression model

```python
model = LinearRegression()
model.fit(Xtrain, Ytrain)
```

Predict on test data

```python
ypred = model.predict(Xtest)
ypred
```

```python
Ytest
```

```python
r2_score(Ytest, ypred)
```
